# Data Assimilation using Optimal Interpolation
Optimal Interpolation is a method where the analysis 
ùê¥ is a weighted average of the model output 
ùëÄ and observations O 

A=M+K(O‚àíM)

where 
K is the Kalman gain factor. For simplicity, we assume 
ùêæ as a constant scalar value, typically chosen between 0 and 1.


```python
import numpy as np

# Sample observational data (in degrees Celsius)
observations = np.array([15.2, 16.1, 14.5, 15.8, 25.0])

# Sample model predictions (in degrees Celsius)
model_output = np.array([14.8, 15.5, 14.0, 16.0, 25.3])

# Kalman gain factor (assumed as a constant scalar for simplicity)
K = 0.5

# Optimal Interpolation (OI)
analysis = model_output + K * (observations - model_output)

# Function to calculate RMSE
def calculate_rmse(predictions, targets):
    return np.sqrt(((predictions - targets) ** 2).mean())

# Function to calculate bias
def calculate_bias(predictions, targets):
    return (predictions - targets).mean()

# Calculate RMSE and bias for model output
rmse_model = calculate_rmse(model_output, observations)
bias_model = calculate_bias(model_output, observations)

# Calculate RMSE and bias for analysis
rmse_analysis = calculate_rmse(analysis, observations)
bias_analysis = calculate_bias(analysis, observations)

# Display results
print("Model Output:", model_output)
print("Observations:", observations)
print("Analysis:", analysis)
print("RMSE (Model):", rmse_model)
print("Bias (Model):", bias_model)
print("RMSE (Analysis):", rmse_analysis)
print("Bias (Analysis):", bias_analysis)
```

# Summary:
Analysis Fields: The analysis fields represent the best estimate of the atmospheric state after combining the observational data with the model output using Optimal Interpolation.
RMSE: The Root Mean Square Error (RMSE) between the analysis fields and the observational data is approximately 0.2114. This indicates the average magnitude of the error.
Bias: The bias between the analysis fields and the observational data is approximately -0.101. This indicates that, on average, the analysis fields slightly underestimate the observational data.
The data assimilation process has improved the estimates by reducing the error and bias compared to the original model output.

```python
import matplotlib.pyplot as plt

# Plotting the observations, model output, and analysis
plt.figure(figsize=(10, 6))
plt.plot(observations, 'o-', label='Observations')
plt.plot(model_output, 's-', label='Model Output')
plt.plot(analysis, 'd-', label='Analysis')
plt.xlabel('Time Step')
plt.ylabel('Temperature (¬∞C)')
plt.title('Observations, Model Output, and Analysis')
plt.legend()
plt.grid(True)
plt.show()
```
